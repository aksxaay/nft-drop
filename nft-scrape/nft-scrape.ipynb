```python
import requests
import json
import os

import threading

# 

def gif_downloader(URI, FILENAME):
  uri = URI
  print(uri)

  filename = FILENAME + '.gif'
  DIR = './tmp'
  fullpath = os.path.join(DIR, filename)

  if os.path.exists(fullpath):
    print(fullpath, 'exists + skipping...')
    print(os.path.getsize(fullpath)/(1024*1024), 'MB')


  else:
    os.makedirs(DIR, exist_ok=True)
    print(FILENAME, 'downloading...')
    response = requests.get(uri)
    print(response.status_code)

    with open(fullpath, 'wb+') as f:
      f.write(response.content)
      print(os.path.getsize(fullpath)/(1024*1024), 'MB')



print('opening key_value_pair.json..')
fp = open('key_value_pair.json')
Dict = json.load(fp)
List = list(Dict.items())
# list of tuple to list of lists
LList = [list(element) for element in List]
for i in range(0, len(LList)):
  LList[i][0] = ''.join(e for e in LList[i][0] if e.isalnum())
print('finally made LList')
print(LList[0:3])

# uri = "https://ipfs.infura.io/ipfs/QmPMai28qR5FQjnePs6u8jH3V8dre29ZhX2h6qTC4aDrD8/image.gif"


for i in range(0, 90, 1):
  print("i:", i)
  try:
    t1 = threading.Thread(target=gif_downloader, args=(LList[i][1], LList[i][0],))
    t2 = threading.Thread(target=gif_downloader, args=(LList[90-i][1], LList[90-i][0],))

    t1.start()
    t2.start()

    t1.join()
    t2.join()
  
  except Exception as e:
    print(e)
  # gif_downloader(LList[i][1], LList[i][0])

```
